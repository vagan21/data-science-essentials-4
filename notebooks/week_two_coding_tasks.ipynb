```python
#Now, import the population by county dataset (population_by_county.csv) into a DataFrame named population
population = pd.read_csv('../data/population_by_county.csv')
population
```

```python
#Merge the physicians DataFrame with the population DataFrame
#Keep only the values for Tennessee
#When you merge, be sure the include both the population and urban columns in the merged results
#Save the result of the merge back to physicians
population = population.loc[population['state'] == 'TENNESSEE']
population['county'] = population['county'].str.replace('County', '')
population
```

```python
#one way to merge 
pcp = pd.merge( left = physicians,
               right = population.loc[population['state']=="TENNESSEE"], on="FIPS")
#another way to merge
physicians = pd.merge(left = physicians,
               right = population.loc[population['state']=="TENNESSEE"][['FIPS', 'population', 'urban']], 
         on="FIPS"
        )
```

```python
populations_physicians_merged = pd.merge(left = physicians, right = population, left_on = 'FIPS', right_on = 'FIPS')
populations_physicians_merged = populations_physicians_merged.drop(columns=['county_y', 'state_y'])
populations_physicians_merged
```

```python
populations_physicians_merged = populations_physicians_merged.rename(columns = {'state_x': 'state', 'county_x': 'county'})
```

```python
populations_physicians_merged
```

```python
#How many Tennessee counties are considered urban?
#38 counties are considered urban; 57 rural 
populations_physicians_merged['urban'].value_counts()
```

```python
#First, counties with fewer than 1500 residents per primary care physician are considered to have an "adequate" supply
#Counties with at least 1500 residents but fewer than 3500 residents per primary care physician are considered to have a "moderately inadequate" supply, 
#and counties with at least 3500 residents per primary care physician are considered to have a "low inadequate" supply
#How many counties in Tennessee are in each group?


populations_physicians_merged['civilians_per_physician'] = populations_physicians_merged['population'] / populations_physicians_merged['primary_care_physicians']
```

```python
populations_physicians_merged
```

```python
#moderately adequate supply
#50 counties with at least 1500 residents but fewer than 3500 residents per primary care physician
populations_physicians_merged.loc[(populations_physicians_merged['civilians_per_physician'] >= 1500) & (populations_physicians_merged['civilians_per_physician'] < 3500)].shape
```

```python
populations_physicians_merged.loc[(populations_physicians_merged['civilians_per_physician'] >= 1500) & (populations_physicians_merged['civilians_per_physician'] < 3500)]['urban'].value_counts()
```

```python
#adequate supply 
#14 counties with fewer than 1500 residents per primary care physician are considered to have an "adequate" supply
populations_physicians_merged.loc[(populations_physicians_merged['civilians_per_physician'] < 1500)].shape
```

```python
populations_physicians_merged.loc[(populations_physicians_merged['civilians_per_physician'] < 1500)]['urban'].value_counts()
```

```python
#low inadequate supply
#31 counties with at least 3500 residents per primary care physician are considered to have a "low inadequate" supply
populations_physicians_merged.loc[(populations_physicians_merged['civilians_per_physician'] >= 3500)].shape
```

```python
populations_physicians_merged.loc[(populations_physicians_merged['civilians_per_physician'] >= 3500)]['urban'].value_counts()
```
